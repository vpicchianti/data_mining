```python
import matplotlib.pyplot as plt
import numpy as np
from colour import Color
from matplotlib import figure
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams['figure.dpi'] = 300
sns.set(font_scale=1.2)

train_acc=[0.63,0.62,0.94,0.68,0.70,0.54,0.93, 0.48, 0.48,0.53,0.40, 0.15]
test_acc= [0.50,0.49,0.49,0.48,0.48, 0.47,0.47,0.43, 0.39, 0.39,0.35, 0.15]

models=['Logi_Sel', 'Logi_Whole','GBM', 'NN_L2', 'SVM_base', 'NN_Drop', 'NN_early_stop','RF_55_feats', 'ADA_tuned_tree','RF_15best_feats', 'ADA_Opt_log', 'Dummy']
green= Color("green")
colors = list(green.range_to(Color("red"),10))
colors = [color.rgb for color in colors]
N=len(models)
ind=np.arange(N)

width = 0.3

plt.bar(ind, train_acc, width, label='Train accuracy', color='red')

plt.bar(ind + width , test_acc, width, label='Test accuracy', color='green')

plt.xlabel('Models', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)

plt.xticks(ind + width /2, models, rotation=90)
plt.yticks(np.arange(0, 1.1,0.1))
plt.legend()
plt.show()






```
